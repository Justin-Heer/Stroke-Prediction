```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
import category_encoders as ce
%config Completer.use_jedi = False
```

Import data

```python
df = pd.read_csv(r'database/healthcare-dataset-stroke-data.csv',index_col=0)
```

init the encoder instance

```python
enc = ce.ordinal.OrdinalEncoder()
```

Encode variables

```python
X = enc.fit_transform(df)
```

Create list of variables to iterate over

```python
var_list = X.columns[:-1]
```

```python
X['bmi'] = X['bmi'].interpolate()
```

Loop over every variable and perform a logistic regression fit and then check the summary to determine if the values are significant

```python
for var in var_list:
    y = X['stroke']
    Xtemp = X[var]
    smlog = sm.Logit(y,sm.add_constant(Xtemp)).fit(maxiter=3000)
    print(f"the summary for {var} is \n\n{smlog.summary()}\n \
    The p values are \n\n{smlog.pvalues}\n")
```

We see that according to the p values for the z test, age, hypertension, heart disease, ever married, work type, residence_type and average glucose level are all significant at the 5% level
